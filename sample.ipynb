```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_dunning_kruger_curve(x):
    """
    This function returns the value of the Dunning Kruger curve at the given x.

    Arguments:
    x -- a numeric value at which the Dunning Kruger curve is evaluated
    """
    return x * (x - 4) * (x - 4) / (x * x + 1)


def plot_dunning_kruger_curve(start_point, current_point):
    """
    This function plots the Dunning Kruger curve and marks the start and current points with arrows.

    Arguments:
    start_point -- the starting point on the curve
    current_point -- the current point on the curve
    """
    x_values = np.arange(0, 10.0, 0.001)
    
    # Plot the Dunning Kruger curve
    plt.plot(x_values, calculate_dunning_kruger_curve(x_values), color="black")
    
    # Plot the arrow from start to current point
    arrow_x_values = np.arange(start_point, current_point, 0.001)
    plt.plot(arrow_x_values - 0.1, calculate_dunning_kruger_curve(arrow_x_values), color="red")
    
    # Plot and annotate start and current points
    for point, label in [(start_point, 'start'), (current_point, 'now')]:
        plt.plot(point, calculate_dunning_kruger_curve(point), ".", markersize=20)
        plt.annotate(label, xy=[point + 0.6, calculate_dunning_kruger_curve(point)])
        
    # Plot arrow at the current point
    plt.annotate('',  
                 xy=[current_point + 0.001, calculate_dunning_kruger_curve(current_point + 0.001)],
                 xytext=[current_point, calculate_dunning_kruger_curve(current_point)],
                 arrowprops=dict(width=1, shrink=0.05, color="red"))

    # Hide axes and labels
    plt.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
    plt.tick_params(bottom=False, left=False, right=False, top=False)
    
    # Show plot
    plt.show()


if __name__=="__main__":
    plot_dunning_kruger_curve(0, 5)
```
